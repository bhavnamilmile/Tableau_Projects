## Clean jets-buyer

```python
df = pd.read_csv('raw/jets-buyer.csv')
```

```python
df
```

```python
data_clean = []
with open('raw/jets-buyer.csv') as f:
    data = f.read()
    data_clean = data.split('ADDR_STATE,% Sales,% Ticket Sold')
     
```

```python
new_data=[]
for line in data_clean:
    line_split = line.split(',,')
    for sp in line_split:
        if len(sp.strip())>0 :
            new_data.append(sp.strip())


```

```python
new_data
```

```python
final_tab = {'Home Team': [], 'Guest Team':[], 'Date':[],'ADDR_STATE':[],'% Sales':[],'% Ticket Sold':[]}
home_team = 'Jets'
# team, date, ADDR_STATE,% Sales,% Ticket Sold
for i in range(0, len(new_data), 2):
    team_date = new_data[i]
    stat = new_data[i+1]

    team, date_str = team_date.split(' at Jets ')
    stat_by_states = stat.split('\n')
    for state_1 in stat_by_states:
        state_name, sale, sold = state_1.split(',')
        sale = float(sale.replace('%', ''))
        sold = float(sold.replace('%', ''))
        final_tab['Home Team'].append(home_team.strip())
        final_tab['Guest Team'].append(team.strip())
        final_tab['Date'].append(date_str)
        final_tab['ADDR_STATE'].append(state_name.strip())
        final_tab['% Sales'].append(sale)
        final_tab['% Ticket Sold'].append(sold)




```

```python
df_final = pd.DataFrame.from_dict(final_tab)
```

```python
df_final
```

```python
df_final.to_csv('NFLTicket_cleaned/cleaned_jets-buyer.csv')
```

## Clean National Average

```python
df = pd.read_csv('raw/national-average.csv')
```

```python
df_tickets = df
```

```python
# Remove 'Tickets' from the 'Genre' column
df_tickets['Genre'] = df_tickets['Genre'].str.replace(' Tickets', '', regex=False)
```

```python
df_tickets.head(4)
```

```python
# Define a function to split the 'Genre' into 'City' and 'Team_Name'
def split_genre(genre):
    # Split by space and handle special cases like 'Green Bay'
    parts = genre.split()

    team_name = parts[-1]
    city = ' '.join(parts[:-1])
    return city, team_name
```

```python
# Apply the function and create new columns for 'City' and 'Team_Name'
df_tickets[['City', 'Team_Name']] = df_tickets.apply(lambda x: pd.Series(split_genre(x['Genre'])), axis=1)

# Display the cleaned DataFrame
df_tickets.head(4) #[['City', 'Team_Name', 'Avg TP, $']]  # Show only relevant columns
```

## clean Average Ticket Price

```python
df= pd.read_csv('raw/2014-average-ticket-price.csv')
```

```python
df_event = df
df_event.head(5)
```

```python
df_event_cleaned = df_event.dropna()
```

```python
#Event,Division,"Avg TP, $"
#Baltimore Ravens at Pittsburgh Steelers Tickets on 02-Nov-2014 (9037819),AFC North,202

# Define a function to split the 'Genre' into 'City' and 'Team_Name'
def split_event(event):
    # if not isinstance(event, str):
    #     return None, None, None
    # Split by space and handle special cases like 'Green Bay'
    # print(event)
    team1, rest_part = event.split(' at ')

    team2, rest_part = rest_part.split(' Tickets ')
    # print(rest_part)
    date= rest_part.split('on ')[1].split(' ')[0]
    return team1, team2, date
```

```python
# Apply the function and create new columns for 'City' and 'Team_Name'
df_event_cleaned[['Team1', 'Team2', 'Date']] = df_event_cleaned.apply(lambda x: pd.Series(split_event(x['Event'])), axis=1)

# Display the cleaned DataFrame
df_event_cleaned.head(4) #[['City', 'Team_Name', 'Avg TP, $']]  # Show only relevant columns
```

```python
df_event_cleaned.to_csv('NFLTicket_cleaned/cleaned_2014-average-ticket-price.csv')
```
